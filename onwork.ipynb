**CHECKING OUTLIERS VIA CODE**

```python
numeric_cols=df.select_dtypes(include=['int64','float64']).columns
outliers_features={}
for col in numeric_cols:
    Q1= df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR= Q3-Q1
    lower= Q1-1.5*IQR
    upper= Q3+1.5*IQR
    if df[(df[col]<lower)|(df[col]>upper)].shape[0]>0:
        outliers_features.append(col)
        print("features containing outliers:\n",outliers_features)
```
